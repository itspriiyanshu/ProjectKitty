```python
import json
import matplotlib.pyplot as plt

# Load training history
with open("training_history.json", "r") as f:
    history = json.load(f)

# Plot Loss
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history["loss"], label="Train Loss", marker='o')
plt.plot(history["val_loss"], label="Val Loss", marker='x')
plt.title("Model Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(history["accuracy"], label="Train Accuracy", marker='o')
plt.plot(history["val_accuracy"], label="Val Accuracy", marker='x')
plt.title("Model Accuracy Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Path to your baseline file
BASELINE_PATH = "baseline_good_curve.json"

# Load baseline curves
with open(BASELINE_PATH, 'r') as f:
    baseline_curves = json.load(f)

# List of features in a readable order
feature_labels = {
    "left_elbow_angle": "Left Elbow Angle",
    "right_elbow_angle": "Right Elbow Angle",
    "shoulder_alignment": "Shoulder Alignment",
    "jump_height": "Jump Height",
    "head_stability": "Head Stability",
    "hip_shoulder_separation": "Hip-Shoulder Separation"
}

# Plotting
plt.figure(figsize=(12, 10))
for i, (key, label) in enumerate(feature_labels.items()):
    plt.subplot(3, 2, i + 1)
    data = baseline_curves.get(key, [])
    if data:
        x = np.arange(len(data))
        plt.plot(x, data, label=label, color='blue')
        plt.title(label)
        plt.xlabel("Time Step")
        plt.ylabel("Value")
        plt.grid(True)
        plt.tight_layout()

plt.suptitle("Baseline Motion Curves for Good Bowling Samples", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
